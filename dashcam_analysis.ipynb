```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
df = pd.read_excel("data/dashcam_data.xlsx")
df['Type'] = df['Type'].str.replace(r'\s+','',regex=True).str.strip()
```

```python
def check_values(df, columns):
    for col in columns:
        print(f"unique value : {df[col].unique()}")
        print(f"Value count : {df[col].value_counts()}")
        print(f"Total unique number : {df[col].nunique()}")
        print(f"Total null value: {df[col].isnull().sum()}")
        print(f"Null data : {df[df[col].isnull()]}")
```

```python
# Display the number of dashcam units returned for repair by models
type_counts = df['Type'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
type_counts.plot(kind='bar', color='green')
plt.title("Dashcam Models  Repair Frequency (July 2023 to July 2025)")
plt.xlabel("Model Type")
plt.ylabel("Number of Repairs")
plt.tight_layout()
plt.show()
```

```python
# Display the number of dashcam units returned for repair by months
months_counts = df['Month'].value_counts().sort_index()
months_counts.index = [calendar.month_abbr[int(m)] for m in months_counts.index]

plt.figure(figsize=(10, 6))
plt.plot(months_counts.index, months_counts.values, marker='o', color='red', linewidth=2)
plt.title("Repairs Per Month")
plt.xlabel("Months")
plt.ylabel("Number of Repairs")
plt.tight_layout()
plt.show()
```
